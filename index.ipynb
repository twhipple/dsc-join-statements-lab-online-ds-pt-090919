## Display the names of all the employees in Boston.
Hint: join the employees and offices tables.

```python
#Your code here
conn = sqlite3.connect('data.sqlite')
cur = conn.cursor()
```

```python
cur.execute("""SELECT * 
               FROM employees 
               JOIN offices 
               ON employees.officeCode = offices.officeCode
               WHERE city = "Boston"
               LIMIT 10;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df.head()
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Look at employees first
cur.execute("""SELECT * 
               FROM employees
               ;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df.head(30)
```

```python
# Then look at offices
cur.execute("""SELECT * 
               FROM offices
               ;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df.head()
```

```python
#The first way I tried.
cur.execute("""SELECT city 
               FROM employees 
               JOIN offices 
               ON employees.officeCode = offices.officeCode
               GROUP BY city
               ;""")
df = pd.DataFrame(cur.fetchall()) 
df.columns = [i[0] for i in cur.description]
df.head(10)
```

```python
# The way to do it with USING().
cur.execute("""SELECT city, COUNT(*)
               FROM employees
               JOIN offices
               USING(officeCode)
               GROUP BY city
               ;""")
df = pd.DataFrame(cur.fetchall()) #Take results and create dataframe
df.columns = [i[0] for i in cur.description]
df.head(10)
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Gave it a try.
cur.execute("""SELECT * 
               FROM products
               JOIN orderdetails
               USING(productCode)
               JOIN orders
               USING(orderNumber)
               JOIN customers
               USING(customerNumber)
               JOIN employees
               GROUP BY lastName
               ;""")
df = pd.DataFrame(cur.fetchall()) #Take results and create dataframe
df.columns = [i[0] for i in cur.description]
df.head(10)
```

```python
#Their code

cur.execute("""SELECT firstName, lastName, productName
               FROM employees e
               JOIN customers c
               ON e.employeeNumber = c.salesRepEmployeeNumber
               JOIN orders o
               USING(customerNumber)
               JOIN orderdetails od
               USING(orderNumber)
               JOIN products p
               USING(productCode)""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df.head()
```

## Level Up: Display the Number of Products each employee has sold

```python
#With Hel

cur.execute("""SELECT firstName, lastName, count(productName) AS total_products_sold
               FROM employees e
               JOIN customers c
               ON e.employeeNumber = c.salesRepEmployeeNumber
               JOIN orders o
               USING(customerNumber)
               JOIN orderdetails od
               USING(orderNumber)
               JOIN products p
               USING(productCode)
               GROUP BY lastName;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df.head()
```
